```python
import networkx as nx
import random
```

```python
n = 12
```

```python
e = n * 2.5
K = nx.gnm_random_graph(n,e)
nx.draw_spring(K, with_labels=True, font_size=15)

sp = nx.average_shortest_path_length(K)
ge = nx.global_efficiency(K)
print("Nodes: " + str(n) +"\nEdges: " + str(e) + "\nAverage Path Length: " + str(sp) + "\nGlobal Efficiency: " + str(ge))
```

The more nodes, the higher required ratio for a given efficiency as shown by the decreasing efficiency at a given ratio as the node count increases.

2.) I believe the more edges the better a network is at transmitting information as the average path length decreases and the efficiency increases meaning that the network can transmit information better because it has a shorter distance to travel.
