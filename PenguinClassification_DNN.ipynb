```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import tflearn
from tflearn.data_utils import load_csv
from tflearn.datasets import titanic
import tensorflow as tf
```

```python
from palmerpenguins import load_penguins
```

```python
penguins = load_penguins()
penguins = penguins.dropna()
penguins
```

```python
penguins["island"].value_counts().to_frame()
```

```python
df = penguins[["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g","sex","island"]]
dict = {'male':0,'female':1}
df = df.replace({'sex':dict})
dict = {'Biscoe':0,'Dream':1, 'Torgersen':2}
df = df.replace({'island':dict})
df = df.to_numpy()
cat = penguins[["species"]]
dict = {'Adelie':0,'Gentoo':1, 'Chinstrap':2}
cat = cat.replace({'species':dict})
cat = cat['species']
cat = cat.to_numpy()

#print(df)
print(cat)
```

```python
alist = []
for i in range(len(cat)):
  i_ind = cat[i]
  tempArray = [0,0,0]
  tempArray[i_ind] = 1
  alist.append(tempArray)
labels = np.array(alist)
labels
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, labels, test_size=0.30, random_state=42)
#input layer
net = tflearn.input_data(shape=[None, np.shape(df)[1]])
#hidden layer 1 with 32 units
net = tflearn.fully_connected(net, 32)
#hidden layer 2 with 32 units
net = tflearn.fully_connected(net, 32)
#output layer
net = tflearn.fully_connected(net, 3, activation='softmax')
```

```python
# Define model
net = tflearn.regression(net)
model = tflearn.DNN(net)
# Start training (apply gradient descent algorithm)
model.fit(X_train, y_train, n_epoch=20, batch_size=32, show_metric=True)
```

```python
prediction_train = pd.DataFrame(model.predict(X_train))
print(prediction_train)
```

```python
predict = prediction_train.idxmax(axis='columns')
prediction_strength = prediction_train.max(axis='columns')
value = pd.DataFrame(y_train).idxmax(axis='columns')
analysisDF = pd.DataFrame(predict,columns=['Prediction'])
analysisDF['Prediction_Strength'] = prediction_strength
analysisDF['Actual_Value'] = value
correct = [1 if predict[x] == value[x] else 0 for x in range(len(predict))]
analysisDF['Correct'] = correct
accuracy = analysisDF['Correct'].sum()/len(value)
print(accuracy)
analysisDF
```

```python
prediction_train = pd.DataFrame(model.predict(X_test))

predict = prediction_train.idxmax(axis='columns')
prediction_strength = prediction_train.max(axis='columns')
value = pd.DataFrame(y_test).idxmax(axis='columns')
analysisDFtest = pd.DataFrame(predict,columns=['Prediction'])
analysisDFtest['Prediction_Strength'] = prediction_strength
analysisDFtest['Actual_Value'] = value
correct = [1 if predict[x] == value[x] else 0 for x in range(len(predict))]
analysisDFtest['Correct'] = correct
accuracy = analysisDFtest['Correct'].sum()/len(value)
print(accuracy)
analysisDFtest
```

```python
metric_train = model.evaluate(X_train, y_train)
metric_test = model.evaluate(X_test, y_test)
print('Model Accuracy on train set: %.9f' % metric_train[0])
print("Model Accuracy on test set: %.9f" % metric_test[0])
```

```python
result = pd.concat([analysisDF, analysisDFtest])
numResult = result
dict = {0:'Adelie',1:'Gentoo',2:'Chinstrap'}
result = result.replace({'Prediction':dict})
result = result.replace({'Actual_Value':dict})
result
```

```python
penCount = result['Actual_Value'].value_counts()

dfValue = result.loc[result['Correct'] == 0]
dfValue
```

```python
penCountError = dfValue["Actual_Value"].value_counts()

tempDF = pd.DataFrame()
tempDF["Amount"] = penCount
tempDF["Accuracy"] = pd.Series(1 - penCountError.div(penCount))
tempDF
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(result.Actual_Value, result.Prediction)
print(cm)

display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels = ['Adelie','Chinstrap','Gentoo'])

plt.rcParams['figure.figsize'] = [15,10]
display.plot()
display.ax_.set_title("DNN Values",fontsize = 50, fontweight = "bold")
display.ax_.set_ylabel("True Label", fontsize = 25, fontweight = "bold")
display.ax_.set_xlabel("Predicted Label", fontsize = 25, fontweight = "bold")
```
